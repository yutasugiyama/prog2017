```python
問106
import random
while True:
   rand_num = random.randint(1,6)
   print(rand_num)
   if rand_num == 4:
    break
   print('.')
```

4が出たらbreakで無限ループを強制終了。

```python
問107
import random
while True:
  rand_num = random.randint(1,6)
  print(rand_num)
  if rand_num == 4:
    continue
  print('.')
```

4が出たら結果を無視して、continueで最初に戻る。

```python
問114
for x in range(1, 21):
    if x % 3 == 0:
        if x % 5 == 0:
            print('Fuzz Buzz')
        else:
            print('Fizz')
    elif x % 5 == 0:
        print('Buzz')
    else:
        print(x)
```

```python
for x in range(1, 21):
    if x % 3 == 0 and x % 5 != 0:
        print('Fizz')
    if x % 3 != 0 and x % 5 == 0:
        print('Buzz')
    if x % 3 == 0 and x % 5 == 0:
        print('Fuzz Buzz')
    if x % 3 != 0 and x % 5 != 0:
        print(x)
```
